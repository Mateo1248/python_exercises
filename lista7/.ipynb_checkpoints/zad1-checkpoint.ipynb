```python
%matplotlib inline
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, display
import numpy as np


def dataCreate(interval: tuple, quantity ,out_func):
    X = np.array([
        np.linspace(
            interval[0],
            interval[1],
            quantity
        )
    ]).T
    Y = out_func(X)
    
    return X, Y    


class Preprocessor:
    def __init__(self, vector):
          self.v_coef = np.max(np.abs(vector))
            
    def scale(self, vector):
        return vector/self.v_coef
    
    def unscale(self, vector):
        return vector*self.v_coef
    

def modelCreate(shape, activation, learning_coef=0.5):
    assert len(shape)-1 == len(activation)
    model = Sequential()
    model.add(
        Dense(
            shape[1],
            input_dim=shape[0],
            activation=activation[0]
        )
    )
    
    for i in range(len(shape)-2):
        model.add(
            Dense(
                shape[i+2],
                activation=activation[i+1]
            )
        )
        
    model.compile(
        optimizer=optimizers.Adam(learning_rate=learning_coef),
        loss='mean_squared_error'
    )
    
    return model

def nn_animation(x, f_x, shape, activation, f_name, step=100, accuracy=2, interval=100, frames=100, learning_coef=0.01):
    
    X, Y = dataCreate(x, (x[1]-x[0]+1)*accuracy, f_x)

    pre_x = Preprocessor(X)
    pre_y = Preprocessor(Y)
    
    x_norm = pre_x.scale(X)
    y_norm = pre_y.scale(Y)

    model = modelCreate(
        shape, 
        activation, 
        learning_coef=learning_coef)

    fig, ax = plt.subplots()
    ax.set_xlim((x[0], x[1]))
    ax.set_ylim((np.min(Y), np.max(Y)))
    line, = ax.plot([], [], lw=2)
    
    def init():
        line.set_data([], [])
        return (line,)
    
    loops = 0
    
    def animate(i):
        model.fit(x_norm, y_norm, epochs=step, verbose=0)
        y_predicted = pre_y.unscale(model.predict(x_norm))  
        line.set_data(X, y_predicted)
        loops = i*step
        plt.title(f"{f_name} loops: {loops}")
        return (line,)   
    
    return FuncAnimation(fig, animate, init_func=init, frames=frames, interval=interval, blit=True)

```

WARSTWY SIECI:





siec 1:10:1

```python
anim = nn_animation(
    (-5,5),
    np.sin,
    [1,10,1], #shape
    ['sigmoid', 'tanh'], #activation
    'sinus',
    accuracy=4,
    step=100,
    interval=150,
    frames=51,
    learning_coef=0.01
)

display(HTML(anim.to_html5_video())) 
```
